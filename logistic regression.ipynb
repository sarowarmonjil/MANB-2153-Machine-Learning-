### Collect Data
#### Download dataset from kaggle. then with pandas, we will read the .CVS

```python
import pandas as pd
import numpy as np
Diabetes=pd.read_csv('diabetes.csv')
table1=np.mean(Diabetes,axis=0)
table2=np.std(Diabetes,axis=0)
```

#### The input and output variables are selected:

```python
inputData=Diabetes.iloc[:,:8]
outputData=Diabetes.iloc[:,8]
```

#### we create and fit a logistic regression model with scikit-learn LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
logit1=LogisticRegression()
logit1.fit(inputData,outputData)
```

#### Model performance

```python
logit1.score(inputData,outputData)
```

```python
import matplotlib.pyplot as plt
plt.figure()
```

#### Visualising Decision boundaries

```python
plt.figure()
plt.scatter(inputData.iloc[:,1],inputData.iloc[:,5],c=logit1.predict_proba(inputData)[:,1],alpha=0.4)
plt.xlabel('Glucose level ')
plt.ylabel('BMI ')
plt.show()

plt.figure()
plt.scatter(inputData.iloc[:,1],inputData.iloc[:,5],c=outputData,alpha=0.4)
plt.xlabel('Glucose level ')
plt.ylabel('BMI ')
plt.show()
```
